```python
# Import pandas for data manipulation
import pandas as pd

# Import matplotlib.pyplot for basic plotting functions
import matplotlib.pyplot as plt

# Import seaborn for advanced/statistical data visualization and loading datasets
import seaborn as sns

# Load the built-in "iris" dataset from seaborn into a pandas DataFrame named 'df'
df = sns.load_dataset("iris")

# Display the first 5 rows of the DataFrame to inspect the data structure
df.head()
```

```python
# Print the dimensions of the dataset (number of rows, number of columns)
df.shape

# Get a list of all column names in the dataset
df.columns.tolist()

# Print a concise summary of the DataFrame, including column names,
# non-null counts, and data types (useful for checking for missing data)
df.info()

# Generate descriptive statistics (mean, standard deviation, min, max, quartiles)
# for all numerical columns in the DataFrame
df.describe()
```

```python
# Define a dictionary mapping each specific Iris species to a color
color_1 = {
    'setosa': 'red',
    'versicolor': 'green',
    'virginica': 'blue'
}

# --- First Scatter Plot: Sepal Measurements ---
# Plot sepal_length on the X-axis and sepal_width on the Y-axis
plt.scatter(
    df['sepal_length'],
    df['sepal_width'],
    # Map the 'species' column to the colors defined in the dictionary above
    c=df['species'].map(color_1)
)
# Add a title and axis labels to the plot
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
# Display the plot
plt.show()

# --- Second Scatter Plot: Petal Measurements ---
# Plot petal_length on the X-axis and petal_width on the Y-axis
plt.scatter(
    df['petal_length'],
    df['petal_width'],
    # Apply the same color mapping to distinguish species
    c=df['species'].map(color_1)
)
# Add a title and axis labels to the plot
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
# Display the plot
plt.show()
```

```python
# --- Sepal Length Histogram ---
# Create a figure with a specific size (10 width, 5 height)
plt.figure(figsize=(10, 5))
# Plot a histogram for sepal_length, adding a KDE (density) curve
sns.histplot(df['sepal_length'], kde=True)
# Add titles and labels
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# --- Sepal Width Histogram ---
plt.figure(figsize=(10, 5))
# Plot a histogram for sepal_width with a KDE curve
sns.histplot(df['sepal_width'], kde=True)
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.show()

# --- Petal Length Histogram (Contains typo in original code) ---
plt.figure(figsize=(10, 5))
# Note: The code plots petal_width here instead of petal_length
sns.histplot(df['petal_width'], kde=True)
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# --- Petal Width Histogram ---
plt.figure(figsize=(10, 5))
# Plot a histogram for petal_width with a KDE curve
sns.histplot(df['petal_width'], kde=True)
plt.title('Distribution of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Define the color mapping again (though it's not strictly used in the loop below)
color_1 = {
    'setosa': 'red',
    'versicolor': 'green',
    'virginica': 'blue'
}

# Create a list containing the names of all the numerical columns we want to plot
numerical_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Loop through each column name in the list
for col in numerical_cols:
    # Set the figure size for each individual plot
    plt.figure(figsize=(10, 6))

    # Create a boxplot:
    # x='species' splits the data into three boxes (one for each species)
    # y=col sets the numerical feature for the y-axis
    # hue='species' colors the boxes based on the species
    sns.boxplot(x='species', y=col, data=df, hue='species')

    # Dynamically set the title and Y-axis label by formatting the column name
    # (e.g., changing 'sepal_length' to 'Sepal Length')
    plt.title(f'Box Plot of {col.replace("_", " ").title()} by Species')
    plt.xlabel('Species')
    plt.ylabel(col.replace("_", " ").title())

    # Add a horizontal grid with dashed lines to make values easier to read
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Display the current plot before moving to the next iteration of the loop
    plt.show()
```
